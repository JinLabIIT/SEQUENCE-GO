```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
data_path = './data/data2.txt'
df = pd.read_csv(data_path, header=None)
df.columns = ['time', 'event_number']
df.insert(2, 'thread_number', 1)
```

```python
df
```

```python
plt.hist(df['event_number'], density=True, bins=10)  # `density=False` would make counts
plt.ylabel('Probability')
plt.xlabel('Number of Event');
```

```python
sample_ratio = 0.01
df_sample = df.sample(frac=sample_ratio)
plt.scatter(df_sample['event_number'], df_sample['time'])
plt.title('sampling = {}'.format(sample_ratio))
plt.ylabel('Time(ns)')
plt.xlabel('Number of Event');
plt.show()
```

```python
data_path = './data/ExceTime.txt'
df = pd.read_csv(data_path, header=None)
df.columns = ['time', 'event_number', 'thread_number']
```

```python
df
```

```python
thread_num = 16
df = df[df['thread_number'] == thread_num]
```

```python
plt.hist(df['event_number'], density=True, bins=10)  # `density=False` would make counts
plt.ylabel('Probability')
plt.xlabel('Number of Event');
```

```python
sample_ratio = 1
df_sample = df.sample(frac=sample_ratio)
plt.scatter(df_sample['event_number'], df_sample['time'], s=1)
plt.title('sampling = {} | thread = {}'.format(sample_ratio, thread_num))
plt.ylabel('Time(ns)')
plt.xlabel('Number of Event');
plt.show()
```
